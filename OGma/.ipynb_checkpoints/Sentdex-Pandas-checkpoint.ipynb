```python
import pandas as pd

df = pd.read_csv("avocado.csv")
df = df.copy()[df["type"] == "organic"]
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by="Date", ascending=True, inplace=True)
```

```python
graph_df = pd.DataFrame()

for region in df['region'].unique():
    print(region)
    
    region_df = df.copy()[df['region'] == region]
    region_df.set_index("Date", inplace=True)
    region_df.sort_index(inplace=True)
    
    region_df[f"{region}_mean25roll"] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[[f"{region}_mean25roll"]]
    else:
        graph_df = graph_df.join(region_df[f"{region}_mean25roll"])
```

```python
graph_df.tail()
```

```python
graph_df.plot()
```

```python
graph_df.plot(figsize=(9,9), legend=False)
```

```python
graph_df.dropna().plot(figsize=(9,9), legend=False)
```

### Groupby - Data Analysis with Python and Pandas p.3

```python
import pandas as pd
df = pd.read_csv("Minimum Wage Data.csv", encoding="latin")
df.head()
```

```python
gd = df.groupby('State')
gd.get_group("Alabama").set_index("Year").head()
```

```python
act_min_wage = pd.DataFrame()

for name, group in df.groupby("State"):
    if act_min_wage.empty:
        act_min_wage = group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name})
    else:
        act_min_wage = act_min_wage.join(group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name}))

act_min_wage.head()
```

```python
act_min_wage.describe()
```

```python
act_min_wage.corr().head()
```

```python
issue_df = df[df["Low.2018"] == 0]
issue_df['State'].unique()
```

```python
import numpy as np
act_min_wage.replace(0, np.NaN).dropna(axis=1).corr().head()
```

```python
issue_df.head()
```

```python
grouped_issues = issue_df.groupby("State")
grouped_issues.get_group("Alabama").head()
```

```python
for state, data in grouped_issues:
    if data['Low.2018'].sum() != 0.0:
        print("We Missed Something")
```
